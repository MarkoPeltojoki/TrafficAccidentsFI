Traffic Accidents Finland 2005-2017

Variables: Age range, Weather, Month and Hour

```python
import requests
import json
from datetime import datetime


# URL for the web service. UusiOnLk v2 with age range, weather, month and hour features.
scoring_uri = 'http://InsertYourOwnEndPointURL.westeurope.azurecontainer.io/score'

# If the service is authenticated, set the key
key = 'InsertYourOwnKey'

# Load some translate values from translate.json file
with open('C:\\InsertYourOwnPath\\translate.json', encoding='utf-8') as tr:
    translate = json.load(tr)

# Input age range as 1-7 (1=0-17, 7=85-), weather as 1-6 (1=clear, 6=slush), month as 1-12, hour as 1-24. Get age, weather and months translated through translate.json file, see also age and weather explanations from there.
agerange = 7
weather = 4
month = datetime.today().month
hour = datetime.today().hour
time = datetime.today().strftime("%H:%M")
agerangetranslate = translate['Age'][str(agerange) + 'a']
weathertranslateFI = translate['Weather'][str(weather) + 'w-fi']
weathertranslateEN = translate['Weather'][str(weather) + 'w-en']
monthtranslateFI = translate['Months'][str(month) + 'm-fi']
monthtranslateEN = translate['Months'][str(month) + 'm-en']

# Data to score for Accident Class
data = {"data":
        [
            [
                agerange,
                weather,
                month,
                hour
            ]
        ]
        }

# Convert to JSON string
input_data = json.dumps(data)

print('Age range, weather, month and hour as Json input data: ' + str(input_data))

# Set the content type
headers = {'Content-Type': 'application/json'}

# If authentication is enabled, set the authorization header
headers['Authorization'] = f'Bearer {key}'

# Make the request and display the response
resp = requests.post(scoring_uri, input_data, headers=headers)
#print(resp.text)
#print(resp.text[16])

if resp.text[16] == "]":
    output_data = (resp.text[15])
else:
    output_data = (resp.text[15:17])

# Create the strings from variables and output data with translations.
# datestring = datetime.today().strftime("%H:%M on %A, %B")
datestringFI =  weathertranslateFI + ', kello ' + time + ' ja kuukausi ' + monthtranslateFI
datestringEN = weathertranslateEN + ' at ' + time + ' on ' + monthtranslateEN
accidentclassEN = translate['Classes'][output_data + 'c-en']
accidentclassFI = translate['Classes'][output_data + 'c-fi']
agestring = agerangetranslate

print('Accident class as output data: '+ str(output_data))
print('Prediction:')
print('Because you are in age range '+ agestring +' and it is '+ datestringEN +', the most probable traffic accident is in class '+ accidentclassEN +'.')
print('Ennuste:')
print('Koska olet ikäluokassa '+ agestring +' ja on '+ datestringFI +', todennäköisin liikenneonnettomuus on luokassa '+ accidentclassFI +'.')
```
